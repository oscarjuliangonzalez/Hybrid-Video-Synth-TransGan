# Train Script Runner

Este notebook ejecuta `train.py` como proceso real para permitir `num_workers > 0` en macOS.

```python
from pathlib import Path
import os
import subprocess
import sys

repo_root = Path.cwd()
script_path = repo_root / 'train.py'

if not script_path.exists():
    raise FileNotFoundError(f'No se encontró {script_path}')

cmd = [
    sys.executable,
    '-u',
    str(script_path),
    '--data-dir', 'EchoNet-Dynamic/sub all frames',
    '--batch-size', '32',
    '--num-workers', '8',
    '--epochs', '100',
    '--save-every', '5',
]

print('Running:', ' '.join(cmd))
env = os.environ.copy()
env['PYTHONUNBUFFERED'] = '1'
process = subprocess.Popen(
    cmd,
    cwd=str(repo_root),
    env=env,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True,
)

for line in process.stdout:
    print(line, end='')

exit_code = process.wait()
print('Exit code:', exit_code)
```
